### 1.1 Clareza

Visualizaes claras permitem que os usurios entendam as informaes rapidamente, sem ambiguidades. Evite sobrecarregar os grficos com elementos desnecessrios.


- Exemplo de m prtica: Grfico confuso

````python
# Dados para o grfico
np.random.seed(10)
data = np.random.rand(10, 4)
categorias = ['A', 'B','C','D']

# M prtica: Grfico confuso com excesso de elementos
plt.figure(figsize=(8, 6))
for i, col in enumerate(categorias):
    print(i,col)
    print(data[:, i])
    plt.plot(data[:, i], label=f"Categoria {col}")
plt.title("Exemplo de Grfico Confuso")
plt.legend()
plt.show()
````

- Exemplo de m prtica: Grfico confuso

O primeiro exemplo demonstra um grfico confuso devido a diversos problemas:

1. **Excesso de Linhas**: O uso de mltiplas linhas sobrepostas dificulta a identificao de padres claros entre as categorias.
2. **Falta de Destaque**: No h distino suficiente entre os dados mais importantes e os elementos complementares.
3. **Legendas Exageradas**: O uso de uma legenda com muitas categorias sobrecarrega o leitor, tornando mais difcil identificar as linhas correspondentes.

Esses problemas podem levar  sobrecarga cognitiva, fazendo com que o pblico tenha dificuldade em interpretar as informaes apresentadas.

---

- Exemplo de boa prtica: Grfico claro

````python

# Boa prtica: Grfico claro
plt.figure(figsize=(8, 6))
valoresMedios = data.mean(axis=0)
plt.bar(categorias, valoresMedios,color=['black', 'gray', 'green', 'red'])
plt.title("Exemplo de Grfico Claro")
plt.ylabel("Valores Mdios")
plt.xlabel("Categorias")
plt.show()
valoresMedios
````

- Exemplo de boa prtica: Grfico claro

No segundo exemplo, o grfico  claro e eficaz pelos seguintes motivos:

1. **Simplicidade**: O grfico apresenta apenas os valores mdios das categorias, removendo informaes redundantes ou desnecessrias.
2. **Uso de Barras**: As barras so mais adequadas para comparar valores entre categorias, tornando as diferenas mais visuais e bvias.
3. **Cores Distintas**: As cores escolhidas ajudam a distinguir as categorias sem causar confuso ou distrao.
4. **Elementos Relevantes**: Apenas os elementos essenciais, como ttulo, eixos e barras, so exibidos, facilitando a compreenso do pblico.

Este grfico prioriza a comunicao eficiente, garantindo que as informaes sejam transmitidas de forma rpida e precisa.

### 1.3 Uso de Cores
As cores devem ser usadas com propsito e moderao. Escolha paletas que sejam acessveis para pessoas com daltonismo e evite combinaes que confundam o leitor.

- Exemplo de m prtica: Uso inadequado de cores

````python

# M prtica: Uso inadequado de cores
categorias = ['Categoria A', 'Categoria B', 'Categoria C', 'Categoria D']
valores = [15, 25, 30, 10]

plt.figure(figsize=(8, 6))
plt.bar(categorias, valores, color=['red', 'green', 'blue', 'yellow'])
plt.title("Grfico com Cores Confusas")
plt.xlabel("Categorias")
plt.ylabel("Valores")
plt.show()
````

- O primeiro exemplo apresenta problemas significativos relacionados ao uso de cores:

1. **Combinaes Distrativas**: O uso de cores como vermelho, verde, azul e amarelo cria uma combinao visualmente confusa e potencialmente problemtica para pessoas com daltonismo.
2. **Falta de Propsito**: As cores no seguem uma lgica ou um esquema claro, tornando difcil entender se h alguma relao entre as categorias representadas.
3. **Esttica Comprometida**: A escolha de cores saturadas e no harmoniosas d ao grfico uma aparncia amadora, reduzindo sua eficcia em ambientes profissionais.

Essas escolhas comprometem a legibilidade e tornam a interpretao do grfico mais difcil, especialmente para pblicos diversos.


````python

# Boa prtica: Uso apropriado de cores
plt.figure(figsize=(8, 6))
sns.barplot(x=categorias, y=valores, palette="viridis")
plt.title("Grfico com Cores Apropriadas")
plt.ylabel("Valores")
plt.xlabel("Categorias")
plt.show()
````

No segundo exemplo, as cores so utilizadas de forma eficaz e intencional:

1. **Paleta Harmoniosa**: O uso de uma paleta consistente e visualmente agradvel (como a "viridis") cria uma experincia visual mais fluida e acessvel.
2. **Clareza e Contraste**: As cores escolhidas oferecem contraste suficiente para diferenciar as categorias sem causar distrao.
3. **Inclusividade**: A paleta foi projetada para ser acessvel a pessoas com daltonismo, garantindo que todos os pblicos possam interpretar o grfico adequadamente.

Este grfico demonstra como o uso estratgico de cores pode melhorar a clareza, a esttica e a acessibilidade, tornando a visualizao mais eficiente e inclusiva para todos os pblicos.

### 2. DashBoards

### Conceito
Um ``dashboard``  uma ferramenta visual que apresenta informaes, dados e mtricas de forma consolidada e interativa.
Ele funciona como um "painel de controle", permitindo que os usurios monitorem indicadores-chave de desempenho (Key Performance Indicator- KPIs) e
tomem decises com base em dados em tempo real.

Dashboards so amplamente utilizados em diversas reas devido  sua capacidade de traduzir dados complexos em visualizaes intuitivas, como grficos, tabelas e mapas. Isso facilita a interpretao de tendncias, padres e insights importantes.

---

### Por que Dashboards so importantes?
Dashboards so fundamentais para:
- ``Monitoramento em tempo real:`` Permitem visualizar o desempenho de processos, vendas ou outros indicadores  medida que os dados so gerados.
- ``Tomada de deciso:`` Apresentam informaes claras e objetivas para suportar decises estratgicas.
- ``Comunicao de dados:`` Facilitam o compartilhamento de resultados com equipes e stakeholders de forma visualmente atraente.
- ``Eficincia operacional:`` Centralizam informaes relevantes em um nico local, reduzindo o tempo gasto na busca e interpretao de dados.

---

### Exemplos Prticos de Uso
1. Empresas monitorando vendas:
   - Acompanhamento de vendas dirias, mensais e anuais.
   - Identificao de produtos ou regies com melhor desempenho.
   - Comparao de receitas e custos para avaliar a lucratividade.

2. Anlise financeira:
   - Controle de despesas e receitas em tempo real.
   - Identificao de tendncias financeiras, como aumentos ou quedas em receitas.
   - Projees de fluxo de caixa e lucros.

3. Controle de operaes:
   - Monitoramento de estoques em tempo real.
   - Anlise de tempo de entrega de pedidos.
   - Deteco de gargalos em processos operacionais.

O layout  definido com os seguintes componentes:
1. **Ttulo:** Um cabealho centralizado (`html.H1`).
2. **Dropdown:** Um menu suspenso para selecionar o continente (`dcc.Dropdown`).
3. **Grfico:** Um espao para exibir grficos interativos (`dcc.Graph`).
4. **Texto Dinmico:** Um espao para mostrar informaes extras.


O callback faz a conexo entre o dropdown e o grfico:
1. **Entrada:** O valor selecionado no dropdown (`dropdown-continent`).
2. **Sadas:**
   - Um grfico de disperso atualizado (`scatter-plot`).
   - Texto mostrando o nmero de pases no continente selecionado (`output-text`).
3. **Lgica:** Filtramos o dataset com base no continente e criamos um grfico atualizado.


Para rodar o aplicativo, usamos:
```python

Isso inicia o servidor local e exibe o dashboard no navegador.

````python
# Importando as bibliotecas necessrias
from dash import Dash, dcc, html, Input, Output
import plotly.express as px
import pandas as pd

# Criando o aplicativo Dash
app = Dash(__name__)

# Criando um dataset para exibio
df = px.data.gapminder()  # Dataset embutido no Plotly

# Layout do aplicativo
app.layout = html.Div([
    html.H1("Dashboard Interativo: Dados de Pases", style={'textAlign': 'center'}),
    dcc.Dropdown(
        id='dropdown-continent',
        options=[
            {'label': continente, 'value': continente}
            for continente in df['continent'].unique()
        ],
        value='Asia',  # Valor padro
        clearable=False,
        style={'width': '50%', 'margin': '0 auto'}
    ),
    dcc.Graph(id='scatter-plot'),
    html.Div(id='output-text', style={'textAlign': 'center', 'marginTop': 20})
])

# Callback para atualizar o grfico e o texto
@app.callback(
    [Output('scatter-plot', 'figure'),
     Output('output-text', 'children')],
    [Input('dropdown-continent', 'value')]
)
def update_dashboard(continente_selecionado):
    df_filtrado = df[df['continent'] == continente_selecionado]
    figura = px.scatter(
        df_filtrado,
        x='gdpPercap',
        y='lifeExp',
        color='country',
        size='pop',
        hover_name='country',
        title=f"Relao entre PIB e Expectativa de Vida ({continente_selecionado})",
        labels={
            'gdpPercap': 'PIB per capita',
            'lifeExp': 'Expectativa de Vida',
            'pop': 'Populao',
            'country': 'Pas'
        }
    )
    texto = f"Total de pases exibidos: {df_filtrado['country'].nunique()}"
    return figura, texto

# Executando o aplicativo
if __name__ == '__main__':
    app.run(debug=True)
````
