```python
# Gerar um conjunto de dados sinttico com 3 clusters
X_sintetico, y_sintetico = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0,n_features=2)

# Converter para DataFrame para facilitar a visualizao
data_sintetico = pd.DataFrame(X_sintetico, columns=['Feature 1', 'Feature 2'])

# Exibir as primeiras linhas do dataset sinttico
data_sintetico.head()
```

### **4. KMeans**


O algoritmo **KMeans**  um dos mtodos mais populares de **clusterizao**. Ele  um algoritmo **no supervisionado** utilizado para particionar um conjunto de dados em **K clusters** baseados nas suas caractersticas. O objetivo do KMeans  dividir os dados de forma que a soma das distncias quadradas entre os pontos e os centrides dos clusters seja minimizada.

**Como funciona o KMeans?**
1. **Inicializao**: O algoritmo comea escolhendo aleatoriamente **K** pontos no espao de caractersticas como os centrides iniciais dos clusters.
2. **Atribuio**: Cada ponto do conjunto de dados  atribudo ao centroide mais prximo, formando **K clusters**.
3. **Atualizao**: Depois que os pontos foram atribudos, o centroide de cada cluster  recalculado como a mdia dos pontos atribudos a ele.
4. **Repetio**: Os passos 2 e 3 so repetidos at que as atribuies de cluster no mudem mais (ou at que o nmero mximo de iteraes seja alcanado).


Agora, vamos aplicar o algoritmo KMeans para agrupar os dados em 3 clusters.



```python

# Aplicar o KMeans com 3 clusters
kmeans = KMeans(n_clusters=3)
data_sintetico['Cluster'] = kmeans.fit_predict(X_sintetico)

# Obter os centrides
centroids = kmeans.cluster_centers_

# Plotar os clusters
plt.figure(figsize=(8, 6))

# Plotar os pontos dos clusters
plt.scatter(data_sintetico['Feature 1'], data_sintetico['Feature 2'], c=data_sintetico['Cluster'], cmap='viridis', s=50, alpha=0.6)

# Plotar os centrides
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X', label='Centrides')

# Ttulos e rtulos
plt.title('Clusterizao com KMeans no Dataset Sinttico')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

```

##### **4.1 Mtodo do Cotovelo (Elbow Method)**

O **Mtodo do Cotovelo**  uma tcnica usada para determinar o nmero ideal de clusters. A ideia  executar o KMeans com diferentes valores de **K** e calcular a **soma das distncias quadradas dentro dos clusters** (tambm conhecida como Inertia). O grfico resultante mostra a relao entre **K** e a Inertia, e o ponto onde a diminuio da Inertia comea a desacelerar (formando um "cotovelo") indica o valor de **K** ideal.



```python

# Lista para armazenar a Inertia para diferentes valores de K
inertia = []

# Testar para diferentes valores de K (de 1 a 10)
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_sintetico)
    inertia.append(kmeans.inertia_)

# Plotar o grfico do mtodo do cotovelo
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Mtodo do Cotovelo')
plt.xlabel('Nmero de Clusters (K)')
plt.ylabel('Inertia')
plt.show()
```




**Explicao do grfico:**
- O grfico exibir a **Inertia** (ou soma das distncias quadradas) para diferentes valores de **K**.
- O **cotovelo** no grfico indica o ponto onde a diminuio da Inertia comea a ser mais lenta. Esse  o valor ideal de **K** para o nosso conjunto de dados.





##### **4.2 ndice de Silhueta (Silhouette Score)**

O **ndice de Silhueta**  uma medida de quo bem os pontos dentro de um cluster esto agrupados em relao aos pontos de outros clusters. O ndice varia de **-1 a 1**:
- **1**: indica que os clusters esto bem separados.
- **0**: indica que os pontos esto no limite entre dois clusters.
- **Valores negativos**: indicam que os pontos esto mal agrupados e podem estar no cluster errado.

A ideia aqui  executar o KMeans para diferentes valores de **K** e calcular o **ndice de silhueta** para cada valor. O valor de **K** com o maior ndice de silhueta ser o ideal.





```python
  # Lista para armazenar os ndices de silhueta
  silhouette_scores = []

  # Testar para diferentes valores de K (de 2 a 10, j que K=1 no faz sentido para silhueta)
  for k in range(2, 11):
      kmeans = KMeans(n_clusters=k)
      kmeans.fit(X_sintetico)
      score = silhouette_score(X_sintetico, kmeans.labels_)
      silhouette_scores.append(score)

  # Plotar o grfico do ndice de silhueta
  plt.figure(figsize=(8, 6))
  plt.plot(range(2, 11), silhouette_scores, marker='o')
  plt.title('ndice de Silhueta para Diferentes Valores de K')
  plt.xlabel('Nmero de Clusters (K)')
  plt.ylabel('Silhouette Score')
  plt.show()
```


**Explicao do grfico:**
- O grfico mostra o **ndice de silhueta** para diferentes valores de **K**.
- O valor de **K** que apresenta o **ndice de silhueta** mais alto indica o nmero ideal de clusters, j que  o valor onde a separao entre os clusters  melhor.


#### **4.3 Interpretao dos Resultados**

1. **Mtodo do Cotovelo**: O grfico do mtodo do cotovelo nos ajuda a identificar o ponto onde o benefcio de adicionar mais clusters comea a diminuir. O "cotovelo" ser o valor de **K** ideal.
   
2. **ndice de Silhueta**: O grfico do ndice de silhueta nos permite avaliar a qualidade da clusterizao. O valor de **K** com o maior ndice de silhueta  o que proporciona a melhor separao entre os clusters.

## **5. Clusterizao com DBSCAN**

O **DBSCAN**  um algoritmo de clusterizao baseado em densidade. Ele identifica clusters como regies densas de pontos separadas por regies de baixa densidade. Essa abordagem permite que o DBSCAN:

- Detecte clusters de formatos arbitrrios.
- Identifique pontos considerados rudos (outliers).
- No exija a especificao do nmero de clusters previamente, diferentemente do KMeans.

##### **Parmetros Principais do DBSCAN**

1. **`eps` (epsilon):** Define o raio mximo de uma regio ao redor de um ponto para que outros pontos sejam considerados vizinhos.
2. **`min_samples`:**  o nmero mnimo de pontos que devem estar dentro do raio `eps` para que uma regio seja considerada um cluster.

##### **Como Funciona**
1. O algoritmo escolhe um ponto inicial e verifica se ele possui ao menos `min_samples` vizinhos dentro de `eps`.
   - Se sim, inicia um novo cluster.
   - Caso contrrio,  considerado rudo.
2. Os vizinhos desse ponto so explorados recursivamente para expandir o cluster.
3. O processo se repete at que todos os pontos sejam processados.

### **Vantagens e Desvantagens**
**Vantagens:**
- Identifica clusters de formas arbitrrias.
- Robusto contra rudo.

**Desvantagens:**
- Sensvel  escolha dos parmetros `eps` e `min_samples`.
- Pode no funcionar bem em datasets com densidades variveis.



```python
# Padronizar os dados
scaler = StandardScaler()
iris_scaled = scaler.fit_transform(data_iris)

# Aplicar DBSCAN
# Configurar eps e min_samples de forma experimental
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(iris_scaled)

# Adicionar os rtulos ao dataset
iris_df = pd.DataFrame(data_iris, columns=iris.feature_names)
iris_df['Cluster'] = labels

# Plotar os clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=iris_df.iloc[:, 0],  # Primeira feature
    y=iris_df.iloc[:, 1],  # Segunda feature
    hue=iris_df['Cluster'],
    palette='viridis',
    style=iris_df['Cluster'],
    legend='full'
)
plt.title("Clusters Identificados pelo DBSCAN - Dataset Iris")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.legend(title="Cluster")
plt.grid()
plt.show()

# Exibir informaes sobre os clusters
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f"Nmero de Clusters: {n_clusters}")
print(f"Nmero de Pontos Considerados Rudo: {n_noise}")
```

### **6. Clusterizao Hierrquica**

A **clusterizao hierrquica**  um mtodo de agrupamento que organiza os dados em uma hierarquia de clusters. Este tipo de clusterizao pode ser representado graficamente por um dendrograma, que ilustra as fuses ou divises de clusters ao longo de diferentes nveis.


##### **Principais Caractersticas**
1. **Dois Tipos de Clusterizao Hierrquica:**
   - **Aglomerativa (bottom-up):**  
     Inicia com cada ponto como um cluster individual e vai combinando-os iterativamente at formar um nico cluster.
   - **Divisiva (top-down):**  
     Comea com todos os pontos em um nico cluster e os divide iterativamente.

2. **Distncia Entre Clusters:**
   - O mtodo de clculo da proximidade entre clusters pode ser feito de vrias formas, como:
     - **Single Linkage:** Distncia mnima entre pontos de dois clusters.
     - **Complete Linkage:** Distncia mxima entre pontos de dois clusters.
     - **Average Linkage:** Mdia das distncias entre todos os pontos de dois clusters.
     - **Ward:** Minimiza a soma dos quadrados das distncias dentro dos clusters.

3. **Dendrograma:**
   - Uma estrutura em forma de rvore que visualiza o processo de clusterizao.
   - Ajuda a decidir o nmero de clusters ao "cortar" a rvore em um nvel especfico.

##### **Vantagens e Desvantagens**
**Vantagens:**
- Adequado para dados pequenos ou moderados.
- Fcil de interpretar visualmente via dendrograma.

**Desvantagens:**
- No  eficiente para grandes datasets devido ao custo computacional elevado.
- Sensvel a outliers, que podem distorcer a estrutura hierrquica.






**6.1  Clusterizao Hierrquica no Dataset Iris**

```python

# Padronizar os dados
scaler = StandardScaler()
iris_scaled = scaler.fit_transform(data_iris)

# Construir o dendrograma
plt.figure(figsize=(10, 7))
linked = linkage(iris_scaled, method='ward')
dendrogram(linked, labels=None, truncate_mode=None)
plt.title("Dendrograma - Clusterizao Hierrquica (Mtodo Ward)")
plt.xlabel("Amostras")
plt.ylabel("Distncia")
plt.show()

# Aplicar Clusterizao Hierrquica (Agglomerativa)
n_clusters = 3  # Definir o nmero de clusters com base no dendrograma
hierarchical_clustering = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
iris_clusters = hierarchical_clustering.fit_predict(iris_scaled)

# Adicionar os clusters ao DataFrame
iris_df = pd.DataFrame(data_iris, columns=iris.feature_names)
iris_df['Cluster'] = iris_clusters

# Exibir os tamanhos de cada cluster
cluster_sizes = iris_df['Cluster'].value_counts()
print("Tamanhos dos clusters:")
print(cluster_sizes)

# Visualizar clusters usando duas features principais
plt.figure(figsize=(8, 6))
colors = ['red', 'blue', 'green']  # Definir cores para os clusters

for cluster in range(n_clusters):
    cluster_points = iris_scaled[iris_clusters == cluster]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1],
                label=f'Cluster {cluster + 1}', color=colors[cluster], s=50)

plt.title("Clusters Identificados - Clusterizao Hierrquica")
plt.xlabel("Feature 1 (normalizada)")
plt.ylabel("Feature 2 (normalizada)")
plt.legend(title="Clusters")
plt.grid()
plt.show()
```

### **7. Comparao de Algoritmos de Clusterizao**

Nesta prtica, faremos uma anlise comparativa entre os algoritmos **KMeans**, **DBSCAN** e **Clusterizao Hierrquica**, utilizando o dataset **make_moons**. Este dataset foi escolhido por sua complexidade moderada: ele apresenta duas regies com formato de lua crescente, que no so facilmente separveis por limites lineares. Esse padro desafia algoritmos como o KMeans, que assume clusters esfricos, tornando-o ideal para observar as diferenas entre os mtodos.
Cada algoritmo de clusterizao tem seus pontos fortes e fracos, que dependem das caractersticas do dataset e da aplicao:

- **KMeans**:
  - Baseia-se na minimizao da varincia dentro dos clusters.
  -  eficiente e amplamente utilizado, mas pode falhar ao lidar com clusters no esfricos ou de tamanhos desiguais.

- **DBSCAN**:
  - Identifica clusters de densidade, permitindo encontrar agrupamentos de formas arbitrrias.
  -  robusto a outliers, mas sensvel  escolha dos parmetros `eps` (tamanho do raio) e `min_samples` (pontos mnimos para formar um cluster).

- **Clusterizao Hierrquica**:
  - Organiza os dados em uma hierarquia de clusters, o que facilita a visualizao via dendrogramas.
  - Pode ser sensvel a outliers e menos eficiente para grandes datasets.



```python
# Gerar o dataset sinttico
n_samples = 500
noise = 0.05
synthetic_data, _ = make_moons(n_samples=n_samples, noise=noise, random_state=42)

# Padronizar os dados
scaler = StandardScaler()
synthetic_scaled = scaler.fit_transform(synthetic_data)

# 1. Aplicar KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_clusters = kmeans.fit_predict(synthetic_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(synthetic_scaled[:, 0], synthetic_scaled[:, 1], c=kmeans_clusters, cmap='viridis', s=10)
plt.title("Clusters Identificados pelo KMeans")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.colorbar(label="Cluster")
plt.grid()
plt.show()

# 2. Aplicar DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan_clusters = dbscan.fit_predict(synthetic_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(synthetic_scaled[:, 0], synthetic_scaled[:, 1], c=dbscan_clusters, cmap='viridis', s=10)
plt.title("Clusters Identificados pelo DBSCAN")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.colorbar(label="Cluster")
plt.grid()
plt.show()

# Determinar o nmero de clusters baseado no dendrograma
hierarchical_clustering = AgglomerativeClustering(n_clusters=2, linkage='ward')
hierarchical_clusters = hierarchical_clustering.fit_predict(synthetic_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(synthetic_scaled[:, 0], synthetic_scaled[:, 1], c=hierarchical_clusters, cmap='viridis', s=10)
plt.title("Clusters Identificados pela Clusterizao Hierrquica")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.colorbar(label="Cluster")
plt.grid()
plt.show()

# Comparao Final
print("Resumo dos resultados:")
print(f"KMeans: {len(set(kmeans_clusters))} clusters")
print(f"DBSCAN: {len(set(dbscan_clusters)) - (1 if -1 in dbscan_clusters else 0)} clusters (incluindo rudo)")
print(f"Clusterizao Hierrquica: {len(set(hierarchical_clusters))} clusters")
```
