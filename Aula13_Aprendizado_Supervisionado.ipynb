#### **O que  Aprendizado Supervisionado?**
O aprendizado supervisionado  um tipo de aprendizado de mquina onde o modelo aprende a partir de dados rotulados. Isso significa que para cada entrada (input) no conjunto de dados, temos uma sada correspondente (label). O objetivo  ensinar o modelo a prever a sada correta para novas entradas baseando-se no que foi aprendido.

Exemplo:  
Imagine que voc tem um conjunto de dados sobre casas, com informaes como tamanho (em m), nmero de quartos e preo de venda. O objetivo do aprendizado supervisionado  criar um modelo que possa prever o preo de uma casa, dado seu tamanho e nmero de quartos.  

#### **O que  Regresso Linear? **
A regresso linear  uma tcnica de aprendizado supervisionado usada para prever valores contnuos. O modelo tenta encontrar a melhor linha reta que se ajusta aos dados, minimizando a distncia entre a linha e os pontos reais.

**Frmula Matemtica da Regresso Linear**:  


$ y = w \cdot x + b $

- \( y \): valor previsto (varivel dependente)  
- \( x \): entrada (varivel independente)  
- \( w \): coeficiente angular (peso do atributo)  
- \( b \): intercepto (constante)

Para medir o acerto da regresso linear usando o dataset sinttico, vamos calcular duas mtricas comuns:

1. **Coeficiente de Determinao (\(R^2\))**: Essa mtrica indica a proporo da variao total da varivel dependente (nota) que  explicada pela regresso.
2. **Erro Quadrtico Mdio (RMSE)**: Essa mtrica quantifica o erro mdio entre as predies e os valores reais.

### Explicao das Mtricas:
1. **$(R^2$)**: O valor varia entre 0 e 1. Quanto mais prximo de 1, melhor o modelo, j que ele consegue explicar a maior parte da variao nos dados.
2. **RMSE**: O RMSE  uma medida da magnitude do erro. Valores menores indicam que as previses esto mais prximas dos valores reais. Ele est na mesma unidade das variveis (neste caso, a nota).

### Exemplo de Sada:
- **$(R^2$)**: Por exemplo, um valor de 0.85 significa que 85% da variao na nota pode ser explicada pelo tempo de estudo.






### **Clculo do Coeficiente de Determinao $R^2$**

A frmula do $R^2$  dada por:

$
R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}
$

Onde:
- $(y_i)$ so os valores reais (notas reais).
- $(\hat{y}_i)$ so os valores previstos.
- $(\bar{y})$  a mdia dos valores reais.

### **Erro Quadrtico Mdio (RMSE)**

A frmula do RMSE :

$
RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
$

Onde:
- $(y_i)$ so os valores reais.
- $(\hat{y}_i)$ so os valores previstos.
- $(n)$  o nmero total de amostras.

Agora, vamos calcular essas mtricas manualmente com base no dataset sinttico.



```python
# Clculo manual de R

# Calcular a mdia dos valores reais (y)
y_media = np.mean(y)

# Calcular o somatrio dos quadrados dos resduos (SQR)
soma_quadrados_residuos = np.sum((y - predicoes) ** 2)

# Calcular o somatrio dos quadrados totais (SQT)
soma_quadrados_totais = np.sum((y - y_media) ** 2)

# Calcular o R manualmente
r2_manual = 1 - (soma_quadrados_residuos / soma_quadrados_totais)
print(f"Coeficiente de Determinao (R) - Clculo Manual: {r2_manual:.4f}")


# Clculo manual de RMSE
# Calcular o erro quadrtico mdio (RMSE)
erro_quadratico_medio = np.sqrt(np.mean((y - predicoes) ** 2))
print(f"Erro Quadrtico Mdio (RMSE) - Clculo Manual: {erro_quadratico_medio:.4f}")
```

### Definindo os Atributos mais Relevantes

O **coeficiente de Lasso** (Least Absolute Shrinkage and Selection Operator)  uma tcnica de regularizao utilizada em modelos de regresso, como a **regresso linear**, para melhorar a generalizao e reduzir o overfitting. O Lasso adiciona um termo de penalizao  funo de custo, que  baseado na soma dos valores absolutos dos coeficientes da regresso. Essa penalizao fora alguns dos coeficientes a se aproximarem de zero, o que pode resultar em um modelo mais simples e com menos variveis significativas.

### A frmula da funo de custo com Lasso:

A funo de custo com regularizao Lasso  dada por:

$
\text{Custo} = \text{MSE} + \lambda \sum_{j=1}^{p} |\beta_j|
$

Onde:
- **MSE**  o erro quadrtico mdio, que mede a diferena entre as predies do modelo e os valores reais.
- **$\beta_j$** so os coeficientes das variveis independentes.
- **$\lambda$**  o parmetro de regularizao que controla a quantidade de penalizao aplicada aos coeficientes.

### O que o Lasso faz:
- Quando **$\lambda$** = 0, o Lasso no penaliza os coeficientes, o que equivale a uma regresso linear simples.
- Quando **$\lambda$** aumenta, o Lasso fora mais coeficientes a se aproximarem de zero. Isso leva a um modelo mais esparso (com menos variveis) e pode ser til para seleo de caractersticas (atributos).

### Vantagens do Lasso:
- **Seleo de Atributos:** O Lasso pode ser usado para selecionar automaticamente as variveis mais relevantes, uma vez que ele tende a reduzir a magnitude dos coeficientes de variveis irrelevantes a zero.
- **Evitar Overfitting:** Ele ajuda a evitar o overfitting, especialmente quando h muitas variveis.

### Usando Matriz de correlao para encontrar os atributos mais reelevantes


A **matriz de correlao**  uma ferramenta estatstica que mede a relao entre diferentes variveis em um conjunto de dados. Ela  especialmente til para entender como as variveis esto relacionadas entre si, o que pode ajudar a identificar quais variveis so mais importantes para modelos preditivos, como a **regresso linear**.

#### Definindo Correlao:

A correlao entre duas variveis pode ser representada por um valor numrico chamado **coeficiente de correlao**, que varia entre -1 e 1:

- **+1**: Correlao positiva perfeita (quando uma varivel aumenta, a outra tambm aumenta de forma proporcional).
- **-1**: Correlao negativa perfeita (quando uma varivel aumenta, a outra diminui de forma proporcional).
- **0**: Nenhuma correlao linear (no h relao linear entre as variveis).

O coeficiente de correlao mais comum  o **coeficiente de correlao de Pearson**, que mede a fora da relao linear entre duas variveis.

#### Matriz de Correlao:

A **matriz de correlao**  uma tabela onde cada clula mostra o valor de correlao entre duas variveis. No caso do nosso dataset de **Boston Housing**, a matriz de correlao ajuda a entender como as variveis, como o nmero de quartos (RM), a taxa de criminalidade (CRIM), a idade das casas (AGE), entre outras, se relacionam entre si.

Uma **alta correlao positiva** (perto de +1) ou **alta correlao negativa** (perto de -1) indica que as variveis esto fortemente relacionadas, enquanto uma **correlao prxima de 0** sugere que no h uma relao linear significativa.

### Como Usar a Matriz de Correlao:

1. **Identificao de Atributos Relevantes**: Podemos identificar quais atributos esto fortemente correlacionados com a varivel dependente (no caso, o preo das casas). Se houver atributos altamente correlacionados, eles podem ser usados para melhorar o modelo.
2. **Reduo de Multicolinearidade**: Quando vrias variveis esto fortemente correlacionadas entre si, isso pode causar problemas em modelos de regresso (multicolinearidade). A matriz de correlao ajuda a identificar e resolver esse problema.

### Visualizando a Matriz de Correlao:

Agora, vamos calcular e plotar a **matriz de correlao** para o dataset **Boston Housing**. Vamos ver como as variveis se relacionam entre si e quais so as mais relevantes para o modelo de regresso.


```python
# Calcular a matriz de correlao
correlacao = boston_df.corr()

# Plotar a matriz de correlao
plt.figure(figsize=(10,8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Matriz de Correlao do Dataset Boston Housing")
plt.show()
```



### **Validao Cruzada**



A **validao cruzada (cross-validation)**  uma tcnica usada para avaliar a performance de um modelo de aprendizado de mquina de maneira mais robusta, ajudando a evitar problemas como **overfitting** ou **underfitting**. A ideia central  dividir o dataset em mltiplos subconjuntos e realizar mltiplos treinos e testes para garantir que o modelo tenha um desempenho consistente em diferentes partes dos dados.

#### **Passos da Validao Cruzada**:
1. **Dividir o dataset**: O conjunto de dados  dividido em "k" subconjuntos (ou "folds"). Por exemplo, se estivermos usando 5-fold cross-validation, o conjunto ser dividido em 5 partes.
2. **Treinar o modelo**: O modelo  treinado em \(k-1\) subconjuntos (dados de treino) e testado no subconjunto restante (dados de teste). Este processo  repetido para cada uma das \(k\) partes.
3. **Mdia das mtricas**: O desempenho do modelo  avaliado pela mdia das mtricas (como \( R^2 \), RMSE) obtidas em cada rodada de validao. Essa mdia fornece uma estimativa mais confivel da capacidade de generalizao do modelo.

#### **Vantagens da Validao Cruzada**:
- **Evita overfitting**: Ao treinar e testar o modelo em diferentes partes do conjunto de dados,  mais difcil que o modelo memorize dados especficos.
- **Avaliao mais robusta**: Ao usar vrias divises dos dados, obtemos uma avaliao mais precisa do desempenho do modelo.

#### **Como Usaremos a Validao Cruzada no Dataset Boston Housing?**
Vamos aplicar a **validao cruzada** utilizando o modelo de **regresso linear** no dataset **Boston Housing**. A mtrica que vamos focar para avaliao ser o **coeficiente de determinao $ R^2$**, que mede o quanto do comportamento dos dados  explicado pelo modelo.
